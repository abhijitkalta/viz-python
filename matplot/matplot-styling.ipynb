```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label="X vs x**2", color='red', marker='+')
axes.plot(x, x, label= "X vs X", color='#e6c017', lw=2, ls='--', marker='o')
axes.legend(loc=(1.1, .5))
fig.suptitle('Legends');
```
