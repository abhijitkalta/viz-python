```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0, 10)
```

```python
y = 2*x
```

```python
plt.plot(x, y);
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(0, 10)
plt.ylim(0, 20)
plt.title('Linear equation');
plt.savefig('firstplot.png')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0, 1, 1])
axes.plot(x, y);
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
fig = plt.figure()
#large axes
axes1 = fig.add_axes([0,0, 1, 1])
axes1.plot(a, b);
axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power')

#small axes
axes2 = fig.add_axes([0.3, 0.5, 0.25, 0.25])
axes2.plot(x, y);
```

```python
fig = plt.figure(figsize=(10, 5), dpi=100)
axes = fig.add_axes([0, 0 , 1, 1])
axes.plot(a, b);
fig.savefig('power.png', bbox_inches='tight')
```
